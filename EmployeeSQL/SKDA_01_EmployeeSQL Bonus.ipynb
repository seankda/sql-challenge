## Prep data for avg salary by title bar chart

```python
# Merge Salaries and Titles Data
bar_combined_data = pd.merge(salaries_data, titles_data, on="emp_no", how="inner")
bar_combined_data.head()
```

```python
# Group by title and get mean
bar_grouped_df = bar_combined_data.groupby("title").mean()
bar_grouped_df
```

```python
# Drop emp_no, reset index
bar_title_salary_df = bar_grouped_df.drop(columns = "emp_no")
bar_title_salary_df = bar_title_salary_df.reset_index()
bar_title_salary_df
```

## Create a bar chart of average salary by title

```python
# Set Labels & Title
plt.ylabel("Employee Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Salary by Title")

# Set x, y axis & tick
x_axis = bar_title_salary_df["title"]
ticks = np.arange(len(x_axis))
y_axis = bar_title_salary_df["salary"]

# Plot bar chart
plt.bar(x_axis, y_axis, align="center", alpha=0.9)
plt.xticks(ticks, x_axis, rotation="vertical")


# Save png
plt.tight_layout()
plt.savefig("./Images/average_salary_by_title")

# Show plot
plt.show()
```
